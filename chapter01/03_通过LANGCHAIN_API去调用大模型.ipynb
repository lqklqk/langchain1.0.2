## 1、通过LangchainAPI调用示例

```python
import os

from langchain.chat_models import init_chat_model
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage
import dotenv
from openai import max_retries
# from sympy.physics.units import temperature

dotenv.load_dotenv()
# 1、获取大模型的实例
model = init_chat_model(
    model=os.getenv("MODEL_NAME"),
    # model_provider="openai",  # 使用openAI的方式调用，自动读取配置文件的OPENAI配置
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL"),
    temperature = 0.5,
)
# 2、构造一个提示词模板或者提示词
messages = [
    SystemMessage(content="你是一个智能助手，可以帮助用户解决问题,你的名字叫做小智"),
    HumanMessage(content="你是谁")
]
# 3、调用model的Invoke方法，获取模型的结果
model.invoke(messages)
```

## 2、init_chat_model相关参数配置


1. model: 模型名称
2. base_url: 端点URL
3. temperature: 控制随机性
4. timeout: 控制超时时间
5. max_tokens: 控制最大输出tokens数量
6. max_retries: 请求重试次数


```python
# 传入max_tokens示例：控制模型总输出长度
model = init_chat_model(
    model="deepseek-chat",
    model_provider="deepseek",
    max_tokens = 10
)
# 2、构造一个提示词模板或者提示词
messages = [
    SystemMessage(content="你是一个智能助手，可以帮助用户解决问题,你的名字叫做小智"),
    HumanMessage(content="你是谁")
]
# 3、调用model的Invoke方法，获取模型的结果
model.invoke(messages)
```

### max_retries参数设置

```python
# timeout示例：由于时间过短，会直接失败，失败后会重试
model = init_chat_model(
    model="deepseek-chat",
    model_provider="deepseek",
    timeout = 0.1,
    max_retries=3
)
# 2、构造一个提示词模板或者提示词
messages = [
    SystemMessage(content="你是一个智能助手，可以帮助用户解决问题,你的名字叫做小智"),
    HumanMessage(content="你是谁")
]
# 3、调用model的Invoke方法，获取模型的结果
model.invoke(messages)
```

## 3、构造消息的几种方式

1、直接使用文本，不使用消息列表

```python
# 直接在invoke方法当中传入文本字符串
model.invoke("你是谁")
```

2、通过AIMessage，HumanMessage等类的方式来去构建消息列表

```python
messages = [
    SystemMessage(content="你是一个智能助手，可以帮助用户解决问题,你的名字叫做小智"),
    HumanMessage(content="你是谁")
]
# 3、调用model的Invoke方法，获取模型的结果
model.invoke(messages)
```

3、通过字典方式构建消息列表，在字典当中定义角色和具体的消息内容，role:system,human,ai

```python
messages=[
    # {"role":"system","content":"你是一个智能助手，名字叫做小智"},
    {"role":"human","content":"你是谁"},
    {"role":"ai","content":"我是小智"},
    {"role":"human","content":"我刚才问你什么问题了"}
]
res =model.invoke(messages)
```

## 4、init_chat_model加载模型的原理：
从langchain对应的模型厂商包（例如langchain_openai, langchain_deepseek）来去构造相关实例，返回
。init_chat_model提供了一种统一的构造模型实例的入口方法


```python
from langchain_openai import ChatOpenAI

from langchain.chat_models import init_chat_model
import dotenv
dotenv.load_dotenv()
client = ChatOpenAI(
    model="gpt-4o-mini"
)

client.invoke("你好")
```

## 5、对话式模型和非对话式的模型
对话式模型：可以和用户进行交互对话
非对话式模型：对用户提供的内容进行续写

```python
from langchain_openai import ChatOpenAI,OpenAI
model = OpenAI(
    model="gpt-4o-mini"
)
```

```python
model.invoke("你好")
```

非对话式模型使用场景：一次性交互的任务，比方说:翻译

```python
model.invoke("帮我把这句话翻译成英文：什么是langchain")
```

非对话式模型和对话式模型的底层原理：langchain调用的OpenAI所提供的两套API。
OpenAI官方对于模型API调用提供了多种不同的方式，在此处就有 /chat/completions 和 /completions.

```python
# 通过调整日志级别，可以看到具体langchain向openai发送的请求信息
import logging
logging.basicConfig(level=logging.DEBUG,format="%(asctime)s %(filename)s %(message)s")
```

```python
# 1、对话式模型
client = ChatOpenAI(
    model="gpt-4o-mini"
)
## 对话式的模型时，调用的是openai所提供的/chat/completions
client.invoke("你好")
```

```python
model = OpenAI(
    model="gpt-4o-mini"
)
# 非对话式模型：langchain调用openai /completions 路径，从而实现了模型是非对话式模型
model.invoke("你好")
```

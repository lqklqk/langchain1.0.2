## 1、使用MCP的好处
1. 假如不用MCP.我们现在使用阿里云百练搜索，我们在代码当中写死参数类型和参数名称，传入的参数有3个。现在代码跑起来没问题。直到，阿里云百练平台搜索工具入参发生调整。我们的代码就可能调不通。
2. 在使用了MCP之后：当阿里云百炼平台搜索工具入参发生调整时，MCP Server能够及时的通知给MCP客户端，MCP客户端能够及时把信息通知给LLM,LLM在构造function calling参数时，就能够按照新的入参来去构建。
3. Server端去定义有哪些工具可以使用，工具的入参是什么等。客户端只需要和Server端建立连接之后，就能够获取这些信息，并且能够及时收到更新的消息。
